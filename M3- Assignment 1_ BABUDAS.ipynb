```python
import pandas as pd
```

```python
c_data = pd.read_excel('/Users/sunayanababudas/Desktop/Crowdfunding_data_1000_projects.xlsx')
```

```python
c_data
```

```python
# Creating new data frame with selected columns
new_columns = c_data[['Project_ID', 'school_city', 'Goal', 'num_donors']]
new_columns
```

```python
# Grouping by school city
grouped = new_columns.groupby(['school_city'], as_index = False).mean()
grouped
```

```python
# Descriptive stats of the grouped data 
grouped.describe()
```

```python
# Creating another data frame
new_columns2 = c_data[['Project_ID', 'school_state']]
new_columns2
```

```python
# Merging both the data frames
merged_data = new_columns.merge(new_columns2, on = 'Project_ID')
merged_data
```

```python
# Getting filtered data 
filtered_data = merged_data[(merged_data['school_state']=='MO') & (merged_data['num_donors']>0)]
filtered_data
```
